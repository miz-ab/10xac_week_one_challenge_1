```python
#number of missing values in our dataset in %
def percent_missing(df):
  # Calculate total number of cells in dataframe
    totalCells = np.product(df.shape)

    # Count number of missing values per column
    missingCount = df.isnull().sum()

    # Calculate total number of missing values
    totalMissing = missingCount.sum()

    # Calculate percentage of missing values
    print("The TellCo user dataset contains", round(((totalMissing/totalCells) * 100), 2), "%", "missing values.")

percent_missing(df)
```

```python
#get the percentege of each columns with missing value
percent_missing = df.isnull().sum() * 100 / len(df)
missing_value_df = pd.DataFrame({'percent_missing': percent_missing})

missing_value_df
```

Task 3. 1 - Aggregate, per customer, the following information (treat missing & outliers by replacing by the mean or the mode of the corresponding variable):


*   Average TCP retransmission
*   Average RTT
*   Handset type
*   Average throughput

```python
#create new data frame called new_net_df 
new_net_df = df[['MSISDN/Number','Bearer Id','Handset Type','TCP DL Retrans. Vol (Bytes)', 'TCP UL Retrans. Vol (Bytes)',\
                         'Avg RTT DL (ms)', 'Avg RTT UL (ms)',\
                         'Avg Bearer TP DL (kbps)', 'Avg Bearer TP UL (kbps)']].copy()
```

```python
#fill every empty columns with mode  
for col in new_net_df.columns:
    if(new_net_df[col].isnull().sum()):
        new_net_df[col] = new_net_df[col].fillna(new_net_df[col].mode()[0])
```

```python
#calculate the average value of TCP retransmission,Average RTT,Average throughput & display head 
new_net_df['Total TCP Retrans'] = new_net_df['TCP DL Retrans. Vol (Bytes)'] +new_net_df['TCP UL Retrans. Vol (Bytes)']
new_net_df['Total Throughput'] = new_net_df['Avg Bearer TP DL (kbps)'] +new_net_df['Avg Bearer TP UL (kbps)']

new_net_df['Total RTT'] = new_net_df['Avg RTT DL (ms)'] + new_net_df['Avg RTT UL (ms)']

new_net_df.head()
```

```python
#calculate the aggregate value 
aggregate = {'Total TCP Retrans':'sum', 'Total Throughput':'sum', 'Total RTT':'sum'}
columns = ['MSISDN/Number','Bearer Id','Handset Type', 'Total TCP Retrans', 'Total Throughput', 'Total RTT']
network_per_user_df = new_net_df[columns].groupby('MSISDN/Number').agg(aggregate).reset_index()
network_per_user_df.head()
```
