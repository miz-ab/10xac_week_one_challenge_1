```python
import pandas as pd
import numpy as np
```

```python
#read the data
pd.set_option('max_column', None)
df = pd.read_csv('drive/MyDrive/Week1_challenge_data_source.csv')
df.head(5)
```

```python
#shape of the df 
df.shape
```

```python
#all columns list
df.columns.tolist()
```

```python
#get number of missing values in each columns
df.isna().sum()
```

```python
#number of missing values in our dataset in %
def percent_missing(df):
  # Calculate total number of cells in dataframe
    totalCells = np.product(df.shape)

    # Count number of missing values per column
    missingCount = df.isnull().sum()

    # Calculate total number of missing values
    totalMissing = missingCount.sum()

    # Calculate percentage of missing values
    print("The TellCo user dataset contains", round(((totalMissing/totalCells) * 100), 2), "%", "missing values.")

percent_missing(df)
```

```python
#get the percentege of each columns with missing value
percent_missing = df.isnull().sum() * 100 / len(df)
missing_value_df = pd.DataFrame({'percent_missing': percent_missing})

missing_value_df
```

```python
#remove some columns, they have 50% and above missing values 
df_clean = df.drop(['TCP DL Retrans. Vol (Bytes)', 'TCP UL Retrans. Vol (Bytes)', 'HTTP DL (Bytes)', 'HTTP UL (Bytes)', 'Nb of sec with 125000B < Vol DL','Nb of sec with 1250B < Vol UL < 6250B','Nb of sec with 31250B < Vol DL < 125000B', 'Nb of sec with 37500B < Vol UL', 'Nb of sec with 6250B < Vol DL < 31250B', 'Nb of sec with 6250B < Vol UL < 37500B'], axis=1)
df_clean.shape
```

```python
#Since IMEI, Bearer, MSISDN/Number Id are fundamentals columns in our data set we cant replace none values with mean or mode instead of that replace with data next to them   
df_clean['Bearer Id'] = df_clean['Bearer Id'].fillna(method='ffill')
df_clean['IMEI'] = df_clean['IMEI'].fillna(method='ffill')
df_clean['MSISDN/Number'] = df_clean['MSISDN/Number'].fillna(method='ffill')
```

```python
#get each IMEI number of Occurence
df_clean['Last Location Name'] = df_clean['Last Location Name'].fillna('not_given')
```

```python
#replace other NAN values with mode 
for column in df_clean.columns:
    df_clean[column].fillna(df_clean[column].mode()[0], inplace=True)
```

```python
#now display the cleaned data frame 
df_clean.head(5)
```

```python
df_clean.groupby('MSISDN/Number').agg({'Social Media DL (Bytes)': ['min', 'max'], 'Social Media UL (Bytes)': ['min','max']}).head(10)
```

```python
#number of xDR sessions for each MSISDN/Number
df_clean.groupby('MSISDN/Number').agg({'Bearer Id':'count'}).head(15)
```

```python
#Session duration
df_clean.groupby('MSISDN/Number').agg({'Dur. (ms)':'sum'}).head(10)
```

```python
#the total download (DL) and upload (UL) data
df_clean.groupby('MSISDN/Number').agg({'Total UL (Bytes)':'sum', 'Total DL (Bytes)':'sum'}).head(10)
```

```python
#the total data volume (in Bytes) during this session for each application
df_clean['Google_total_data_volume'] = df_clean['Google DL (Bytes)'] + df_clean['Google UL (Bytes)']
df_clean['Social Media_data_volume'] = df_clean['Social Media DL (Bytes)'] + df_clean['Social Media UL (Bytes)']
df_clean['Email_total_data_volume'] = df_clean['Email DL (Bytes)'] + df_clean['Email UL (Bytes)']
df_clean['Netflix_total_data_volume'] = df_clean['Netflix DL (Bytes)'] + df_clean['Netflix UL (Bytes)']
df_clean['Gaming_total_data_volume'] = df_clean['Gaming DL (Bytes)'] + df_clean['Gaming UL (Bytes)']
df_clean['Other_total_data_volume'] = df_clean['Other DL (Bytes)'] + df_clean['Other UL (Bytes)']
df_clean['Youtube_total_data_volume'] = df_clean['Youtube DL (Bytes)'] + df_clean['Youtube UL (Bytes)']
df_clean['Total_total_data_volume'] = df_clean['Total DL (Bytes)'] + df_clean['Total UL (Bytes)']
df_clean.head(5)
```
